```python
import sys
```

```python
a = 2
sys.getsizeof(a)    # bytes
```

```python
lst = [True] * 100
sys.getsizeof(lst) / len(lst)     # bytes
```

```python
lst = bytearray([True] * 10000)
sys.getsizeof(lst) / len(lst)
```

```python
bits = 0b111111111111111111111111111111
print(sys.getsizeof(bits))
print(bits.bit_length())
```

```python
import sys

for n in [10, 100, 1_000, 10_000]:
    x = (1 << n) - 1
    print(n, sys.getsizeof(x) / n)

```

```python
# create n bits all set to 1
bitset = (1 << 10_000) - 1
sys.getsizeof(bitset) / 10_000
```
